```python
#suppress warnings

import warnings 
warnings.filterwarnings('ignore')

#import necessary libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 


#reading the dataset
life = pd.read_csv('Assignment1_Life_Expectancy_Data.csv')

#remove duplicates 
life_new = life.drop_duplicates()
print('Remaining values in dataset are', life_new.shape)

#print sample data
life_new.head(10)
```

1. Which variables have outliers? (2 Marks)

```python

# Create subplots dynamically based on the number of columns
num_cols = len(life_new.columns)
num_rows = (num_cols + 1) // 2  # Ensures we have enough rows for all columns

fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(12, 5 * num_rows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each column
for i, column in enumerate(life_new.columns):
    sns.boxplot(x=life_new[column], ax=axes[i], orient='h')
    axes[i].set_title(column)

# Adjust layout
plt.tight_layout()
plt.show()


```

```python

# Aggregating data by 'Status' and calculate the sum of 'Life expectancy'
status_life_expectancy = life_new.groupby('Status')['Life expectancy '].sum().reset_index()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the stacked bar chart
sns.barplot(x='Status', y='Life expectancy ', data=status_life_expectancy, ci=None)

# Add title and labels
plt.title('Stacked Bar Chart of Life Expectancy by Status')
plt.xlabel('Status')
plt.ylabel('Total Life Expectancy')

# Show the plot
plt.show()

```

**6.** Which factors are the biggest drivers of Adult mortality? (5 Marks)

```python
# Calculating correlation between Adult Mortality and other variables
# Selecting columns and excluding Adult Mortality
columns_of_interest = ['Life expectancy ', 'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B', 
                      'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure', 'Diphtheria ', 
                      ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years', ' thinness 5-9 years', 
                      'Income composition of resources', 'Schooling', 'Adult Mortality']

# Calculate correlation between Adult Mortality and selected variables
correlation_with_adult_mortality = uganda_data[columns_of_interest].corr()['Adult Mortality'].sort_values(ascending=False)

# Remove Adult Mortality from the correlation series
correlation_with_adult_mortality = correlation_with_adult_mortality.drop('Adult Mortality')

# Plot correlation coefficients
plt.figure(figsize=(10, 8))
sns.barplot(x=correlation_with_adult_mortality.values, y=correlation_with_adult_mortality.index, palette='turbo')
plt.title('Correlation with Adult Mortality')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Variable')
plt.show()
```

*Answer* The 5 biggest factors for Adult Mortality rate in Uganda is Schooling, percentage expendityure, Polio, infant deaths and under five deaths
